## Rolling and Plotting

Let us try to plot the trajectory of this poor $c^k$

We will need a special library for that `Plots`. First, we need to install that with `Pkg`, then build it and use!
Also, we will need `LaTeXStrings` to annotate in a more nifty way.

```julia
#Install `Plots`
Pkg.add("Plots")
Pkg.build("Plots")
using Plots

#Install `LaTeXStrings`
Pkg.add("LaTeXStrings")
Pkg.build("LaTeXStrings")
using LaTeXStrings
```

```julia
c = 0.5 + 0.5 * im
history = []
for k ∈ 1:500
    c_k = power(c, k)
    append!(history, c_k)
end
```

```julia
# Let us see what is inside of that...
println("The history containts:\n", history)
```

### Let us plot it!

```julia
# First, we need to separate real parts from imaginary parts
real_parts = []
imag_parts = []
for c_number in history
    #plot_obj = plot([real(c_number)], [imag(c_number)], seriestype = :scatter)
    append!(real_parts, real(c_number))
    append!(imag_parts, imag(c_number))
end
```

```julia
# Now plot this stuff
#You can plot something
plot(real_parts, imag_parts, label=L"Trajectory ~of~ c^k")
#If you want to amend the existing plot, use `!`, like below
plot!(xlabel="Real")
plot!(ylabel="Imaginary")
```

# Make a nice function
Now let us make a nice function that will make this plot in one line and save to your computer, huh?

```julia
function plot_power_trajectory(c, N)
    """
    Takes complex number c and nubmer of trajectory steps, plots the trajectory and saves it as an image
    c (complex): a number to be powered up
    N     (int): a number of points in the trajectory
    """
    #get the trajectory
    history = []
    for k ∈ 1:N
        c_k = power(c, k)
        append!(history, c_k)
    end
    real_parts = []
    imag_parts = []
    for c_number in history
        #plot_obj = plot([real(c_number)], [imag(c_number)], seriestype = :scatter)
        append!(real_parts, real(c_number))
        append!(imag_parts, imag(c_number))
    end
    plot(real_parts, imag_parts, label=L"Trajectory ~of~ c^k")
    #If you want to amend the existing plot, use `!`, like below
    plot!(xlabel="Real")
    plot!(ylabel="Imaginary")
    savefig("fancy_plot.png")
end
```

```julia
plot_power_trajectory(0.5 + 0.5im, 500)
```
